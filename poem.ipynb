```python
from urllib.request import urlretrieve
url = "https://github.com/Elwing-Chou/ml0602/raw/main/poem_test.csv"
urlretrieve(url, "test.csv")
url = "https://github.com/Elwing-Chou/ml0602/raw/main/poem_train.csv"
urlretrieve(url, "train.csv")
```

```python
import pandas as pd
train = pd.read_csv("train.csv", encoding="utf-8")
test = pd.read_csv("test.csv", encoding="utf-8")
```

```python
# 分類: 輸出 0 1 2 3
# series.replace(字典)
# {"李白":0, "杜甫:":1, "白居易":2}
# series.unique/series.value_counts
u = train["作者"].unique()
print(u)
# name2index = {}
name2index = {n:i for i,n in enumerate(u)} # enmerate = index no.+ items
index2name = {i:n for i,n in enumerate(u)} # n:i 順便把東西放進去 dict
# for i in range(len(u)):
# for i, n in enumerate(u):
  # name2index[n] = i
print(name2index)
print(index2name)
y_train = train["作者"].replace(name2index)
y_test = test["作者"].replace(name2index)
y_test
```

```python
# series.apply(func不加小括號)!!!!!!!!!!!
# !!老師做過實驗, 詩詞分類載入大辭典反而適得其反, 但是普通時候你要載
import jieba
def poemcut(p):
  return " ".join(jieba.cut(p))
x_train = train["內容"].apply(poemcut)
x_test = test["內容"].apply(poemcut)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# 如果出現訓練資料以外的詞就 pass 掉
vec = CountVectorizer()
vec.fit(x_train)
vec.vocabulary_
```

```python
# fit_transform: 測試資料只做transform
# 稀疏矩陣
x_train_count = vec.transform(x_train) # 可以print看, 儲存非0的欄位
x_test_count = vec.transform(x_test)
x_test_count
```

```python
# 檢查標點符號 , \r \n
# vec.vocabulary_["。"]
# vec.vocabulary_["\r\n"]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB(alpha=0.01)
clf.fit(x_train_count, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_count)
accuracy_score(pre, y_test)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre) # "混淆矩陣"可以確認使用結果 --> 看看 data 有無問題
pd.DataFrame(mat,
      columns=["{}(預測)".format(n) for n in u],
      index=["{}(正確)".format(n) for n in u])
```

```python
poem = input("輸入詩:")
poems = [poemcut(poem)]
poems_count = vec.transform(poems)
pre = clf.predict(poems_count)[0]
print("答案是:", index2name[pre])
pre_proba = clf.predict_proba(poems_count)[0]
pre_proba
for name, proba in zip(u, pre_proba):
  print(name,  "的機率:", proba)
```
